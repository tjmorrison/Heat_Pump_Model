```python

'''---------------------------------------------IMPORT PACKAGES -----------------------------------------------------'''
import matplotlib.pyplot as plt

'''----------------------------------------------PLOT SETTINGS ------------------------------------------------------'''
plt.style.use('seaborn-colorblind') # Color Scheme for plots
plt.rcParams["font.family"] = "Times New Roman" # Font style
plt.rcParams.update({'font.size': 22}) #Set font size
```

```python
#Make spatial improvment figure
plt.figure(figsize=(10, 6))
plt.errorbar([2], [73], yerr=15, fmt='bo',capsize=3)
plt.errorbar([5000], [81], yerr=10, fmt='ro',capsize=3)
#plt.title('WBA ' + titles[cnt] + ' Daily Total')
plt.minorticks_on()
plt.grid(b=bool, which='major', axis='both')
plt.xscale('log')
plt.ylabel('Accuracy [%]')
plt.xlabel('Forecasted Resolution Space [Acres]')
plt.ylim((0,100))
plt.xlim((1,10000))
plt.autoscale(enable=True, axis='x', tight=False)

plt.legend(['ALPINE3D','Forecaster'], loc=4) # legend in upper left side
plt.savefig('../gen/Spatial_improvements' +'.png', bbox_inches='tight')
#plt.show()
#plt.close()
```
